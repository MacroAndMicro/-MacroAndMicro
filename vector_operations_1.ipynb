```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0, 0, 4, 5)
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0, 4, -5, scale_units='xy', angles='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

### Addition of vectors

```python
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,-4,5])
v12 = v1+v2
print(v12)
```

```python
plt.quiver(v1[0], v1[1], v1[2], v1[3], scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(v2[0], v2[1], v2[2], v2[3], scale_units='xy', angles='xy', scale=1)
plt.quiver(v12[0], v12[1], v12[2], v12[3], scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

### Subtraction of vectors

```python
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,-4,5])
v12 = v1-v2
print(v12)
```

```python
plt.quiver(v1[0], v1[1], v1[2], v1[3], scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(v2[0], v2[1], v2[2], v2[3], scale_units='xy', angles='xy', scale=1)
plt.quiver(v12[0], v12[1], v12[2], v12[3], scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

### Multiplication of vectors

```python
v1 = np.asarray([0,0,2,3])
v2 = -0.8*v1
v12 = v1+v2
print(v12)
```

```python
plt.quiver(v1[0], v1[1], v1[2], v1[3], scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(v2[0], v2[1], v2[2], v2[3], scale_units='xy', angles='xy', scale=1)
plt.quiver(v12[0], v12[1], v12[2], v12[3], scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```
